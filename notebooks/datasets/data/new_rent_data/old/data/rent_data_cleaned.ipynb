```python
import numpy as np
import pandas as pd
```

```python
# load the data
df_data = pd.read_csv('data/Metro_ZORI_AllHomesPlusMultifamily_Smoothed.csv')
df_data.T
```

```python
# https://www.geeksforgeeks.org/split-a-text-column-into-two-columns-in-pandas-dataframe/

# split the 'RegionName' column by commas
df_split = pd.DataFrame(df_data.RegionName.str.split(',').tolist(), 
                         columns = ['City','State'])

df_split
```

```python
# combine columns from _split to _data
df_data['City'] = df_split['City']
df_data['State'] = df_split['State']
df_data
```

```python
# reduce columns to only return the most recent data
df_cleaned = df_data[df_data.columns[[87, 88, 2, 86]]]
df_cleaned
```

```python
# display the cities that are combined
for x in df_cleaned['City']:
    if x.count('-') > 0:
        print(x)
```

```python
# create a new list containing the required amount of cities to duplicate
dups = []

for i,x in enumerate(df_cleaned['City']):
    if x.count('-') > 0:
        for y in range(0, x.count('-')):
            dups.append(df_cleaned[i:i+1])

# add the duplicated cities to the dataset
for x in dups:
    df_cleaned = df_cleaned.append(x)   
    
df_cleaned
```

```python
# rearrange the dataset by 'SizeRank' so all the duplicates are in 
# the appropriate place
df_cleaned = df_cleaned.sort_values(by='SizeRank') 
df_cleaned
```

```python
# reset index so there are no duplicate indices since that creates 
# complications when parsing through data for additional cleaning
df_cleaned = df_cleaned.reset_index(drop=True)
df_cleaned
```

```python
# create lists which contain the i+n'th combined city and their index. 
# the i'th combined city will remain unchanged but can easily be dealt
# with later
dups_city = []
dups_ind  = []

lst = df_cleaned['City'].tolist()

for i,x in enumerate(lst):
    if i+1 < len(lst):
        x1 = lst[i]
        x2 = lst[i+1]
        if x1 == x2:
            dups_city.append(x)
            dups_ind.append(i)
    else:
        pass
            
print(dups_city)
print()
print(dups_ind)
```

```python
# this code is written specifically for the case that there are at most 
# 3 cities combined. again, the i'th city will remain unchanged for now
dups_final = []

for i,x in enumerate(dups_city):
    if i + 1 < len(dups_city):
        if dups_city[i] == dups_city[i+1]:
            dups_final.append(dups_city[i].split('-')[2])
        else:
            dups_final.append(dups_city[i].split('-')[1])
    else:
        dups_final.append(dups_city[i].split('-')[1])

print(dups_final)
print()
print(dups_ind)
```

```python
# cast the dataframes list of cities to a list and replace its duplicated 
# cities with the city/index lists that we created above
new_cities = df_cleaned['City'].tolist()

for x,y in zip(dups_ind, dups_final):
    new_cities[x] = y
    
new_cities
```

```python
# the remaining i'th cities that are combined will be # split here by 
# taking a string split and replacing it with the 0'th element
for i,x in enumerate(new_cities):
    if len(x.split('-')) > 1:
        new_cities[i] = x.split('-')[0]
    
new_cities
```

```python
# apply the new list of cities to the dataframe
df_cleaned['City'] = new_cities
df_cleaned
```
